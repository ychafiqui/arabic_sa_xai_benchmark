```python
import pandas as pd
from transformers import pipeline
import torch
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import time
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
df = pd.read_csv('TripAdvisor hotel reviews.csv')
```

```python
df.head()
```

```python
df['polarity'].value_counts()
```

```python
df['polarity'] = df['polarity'].replace(0, 'neutral')
df['polarity'] = df['polarity'].replace(1, 'positive')
df['polarity'] = df['polarity'].replace(-1, 'negative')
```

```python
model_name = "CAMeL-Lab/bert-base-arabic-camelbert-msa-sentiment"
pipe = pipeline(model=model_name, device=device, max_length=512, truncation=True)
```

```python
preds = df['text'].apply(lambda x: pipe(x)[0]["label"])
```

```python
accuracy = accuracy_score(df['polarity'], preds)
f1 = f1_score(df['polarity'], preds, average='weighted')
precision = precision_score(df['polarity'], preds, average='weighted')
recall = recall_score(df['polarity'], preds, average='weighted')
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1: {f1}")
```

```python
model_name = "tabularisai/multilingual-sentiment-analysis"
pipe = pipeline(model=model_name, device=device, max_length=512, truncation=True)
```

```python
def predict(text):
  sentiment_map = {0: "negative", 1: "negative", 2: "neutral", 3: "positive", 4: "positive"}
  label = pipe(text)[0]["label"]
  id = pipe.model.config.label2id[label]
  return sentiment_map[id]
```

```python
preds = df['text'].apply(lambda x: predict(x))
```

```python
accuracy = accuracy_score(df['polarity'], preds)
f1 = f1_score(df['polarity'], preds, average='weighted')
precision = precision_score(df['polarity'], preds, average='weighted')
recall = recall_score(df['polarity'], preds, average='weighted')
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1: {f1}")
```
